### Tarea. Generar salida para el archivo bodega

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
data = pd.read_excel('.//documentos-iniciales/informacion-general.xlsx','BODEGAS EP PEC')
```

```python
data.shape
```

```python
data
```

```python
latitud=[]
longitud=[]
```

```python
nombre=data['Unnamed: 0'].tolist()[3:]
codigo=data['Unnamed: 1'].tolist()[3:]
direccion=data['Unnamed: 3'].tolist()[3:]
activainactiva=[]
administrador=data['Unnamed: 2'].tolist()[3:]
correo=data['Unnamed: 4'].tolist()[3:]
telefono=data['Unnamed: 5'].tolist()[3:]
extension=data['Unnamed: 6'].tolist()[3:]
celular=data['Unnamed: 7'].tolist()[3:]
```

```python
salida1=pd.DataFrame([nombre,codigo,direccion,activainactiva,administrador,correo,telefono,extension,celular,latitud,longitud])
```

```python
salida1
salida2=salida1.T
salida2.columns=["nombre","codigo","direccion","activainactiva","administrador","correo","telefono","ext","celular","latitud","longitud"]
```

```python
salida2
```

```python
import re
geolocator = Nominatim()
for i in range(0,14):

    stringuno=re.split('[:().]',salida2['direccion'][i])[0]
    try:
        latitud.append(geolocator.geocode(stringuno).latitude)
        
    except :
        latitud.append(geolocator.geocode("Ecuador").latitude)
        
    
    try:
        longitud.append(geolocator.geocode(stringuno).longitude)
        
    except :
        longitud.append(geolocator.geocode("Ecuador").longitude)
        


```

```python
salida2['latitud']=latitud
salida2['longitud']=longitud
```

```python
salida2
```

```python
salida2['correo']=salida2['correo'].str.lower()
salida2['nombre']=salida2['nombre'].str.upper()
salida2['direccion']=salida2['direccion'].str.upper()
salida2['administrador']=salida2['administrador'].str.upper()
```

```python
def modificarcelular(celular):
    try:
        return '0'+str(int(round(celular,0)))
    except :
        return ''
```

```python
salida2['celular']=list(map(modificarcelular,salida2['celular'].values))
```

```python
salida2['correo']=[re.sub(r'\.ec','.com',item) for item in salida2['correo']]
```

```python
salida2['ext']=[re.sub(r'\n','/',str(item)) for item in salida2['ext']]
```

```python
salida2['activainactiva']=''
```

```python
salida2['telefono'][8]=''
```

```python

salida2.to_excel(".//documentos-finales/bodegafinalpython.xlsx",sheet_name='Bodegas')
```
