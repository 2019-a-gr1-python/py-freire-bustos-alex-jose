```python
import networkx as nx
import scipy
import scipy.optimize
import scipy.stats
import matplotlib.pyplot as plt
import json
import heapq
%matplotlib inline
```

```python
G = nx.barabasi_albert_graph(20, 3, 1)
pos = nx.spring_layout(G)

plt.figure(figsize=(5,5))
nx.draw(G,pos=pos,with_labels=True,node_size=400,cmap=plt.cm.Blues)
plt.show()
```

```python
degree = [G.degree(x) for x in G.nodes()]
label = [x for x in G.nodes()]
h = nx.harmonic_centrality(G)
harmonic = [h[x] for x in G.nodes()]
b = nx.betweenness_centrality(G)
betweenness = [b[x] for x in b]
e = nx.eigenvector_centrality(G)
eigen = [e[x] for x in G.nodes()]
pr = nx.pagerank(G)
pagerank = [pr[x] for x in G.nodes()]

print "Degree:"
print degree
print "PageRank:"
print pagerank
print "Harmonic centrality:"
print h

plt.figure(figsize=(4,4))
nx.draw(G,pos=pos,node_color=harmonic,with_labels=True,node_size=400,cmap=plt.cm.Blues)
plt.show()

```
