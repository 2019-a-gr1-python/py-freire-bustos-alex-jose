```python
import json as json
import numpy as np
```

```python
path="./data/db.json"
db={
    'clientes':[]
}
```

```python
def load_database():
    global db, path
    try:
        with open(path,encoding="utf-8") as json_file:  
            db = json.load(json_file)          
            json_file.close()
    except:
        print("Error loading database.")
```

```python
def save_database():
    global db, path
    try:
        with open(path,'w',encoding="utf-8") as json_file:
            json.dump(db,json_file)
            json_file.close()
        return True
    except:
        print("Error saving database.")
        return False
```

```python
class Cliente:
    numero_socio=''
    nombre=''
    apellido=''
    direccion=''
    cuentas_disponibles=''
    
    def __init__(self,numero_socio,nombre,apellido,direccion,cuentas_disponibles):
        self.numero_socio=numero_socio
        self.nombre=nombre
        self.apellido=apellido
        self.direccion=direccion
        self.cuentas_disponibles=cuentas_disponibles
        
    def __str__(self):
        return (f'Numero de socio: {self.numero_socio}\nNombre: {self.nombre}\nApellido: {self.apellido}\nDireccion: {self.direccion}\nCuentas Disponibles: {self.cuentas_disponibles}')
    
    # def add_account(self,nombre,numero):
      #  registro={'nombre':nombre,'numero':numero}
       # self.cuentas_disponibles.append(registro)
        
    #def __get_bank_accounts(self):
     #   return json.dumps(self.cuentas_disponibles)
    
    def to_dictionary(self):
        cliente={
            'numero_socio':self.numero_socio,
            'nombre':self.nombre,
            'apellido':self.apellido,
            'direccion':self.direccion,
            'cuentas_disponibles':self.cuentas_disponibles
        }
        return cliente
```

```python
def main():
    load_database()
    open_main_menu()
```

```python
def open_main_menu():
    flag_main=1
    while flag_main:
        print("** Gestión de socios - Banco ABC **\nSeleccione una opción:\n1. Gestion de clientes\n0. Salir\n")
        opcion=input()
        if(opcion!='1' and opcion!='2' and opcion!='0'):
            print("\n**Valor invalido**\n")
        else:
            flag_main=0
        if(opcion=='1'):
            manage_clients()
        if(opcion=='0'):
            print("Salio del sistema\n")
```

```python
def manage_clients():
    flag_client=1
    while flag_client:
        print("\n** Gestion de clientes **\n",
                 "Seleccione una opción:\n",
                 "1. Obtener listado de clientes\n",
                 "2. Registrar nuevo cliente\n",
                 "3. Consultar cliente\n",
                 "4. Actualizar cliente\n",
                 "5. Eliminar cliente\n",
                 "0. Atras\n")
        opcion=input()
        if(opcion!='1' and opcion!='2' and opcion!='3'and opcion!='4' and opcion!='5'and opcion!='6' and opcion!='0'):
            print("\n**Valor invalido**\n")
        if(opcion=='1'):
            get_client()
        if(opcion=='2'):
            create_client()
        if(opcion=='3'):
            codigo=input("\nIngrese el código del cliente:\n")
            find_client(codigo)
        if(opcion=='4'):
            update_client()
        if(opcion=='5'):
            delete_client()
        if(opcion=='0'):
            open_main_menu()
            flag_client=0
            
def get_client():
    clientes=db['clientes']
    if(len(clientes)==0):
        print("\n** Cliente no registrado **\n")
        return False
    else:
        print("\nLista de clientes:")    
        for cliente in clientes:
            print("- Código: ",cliente['numero_socio'],"\tNombre: ",cliente['nombre'],"\tApellido: ",cliente['apellido'],"\tDireccion: ",cliente['direccion'], "\tCuentas Disponibles: ",cliente['cuentas_disponibles'])
        return True
    
def create_client():
    cliente_string=input("\nIngrese nuevos datos del cliente \n(Código,Nombre,Apellido,Direccion,CuentaDisponible)\n")
    cliente_array=cliente_string.split(',')
    try:
        cliente=Cliente(cliente_array[0],cliente_array[1],cliente_array[2],cliente_array[3],cliente_array[4])
        index=existe_cliente(cliente.numero_socio)
        if(index==-1):
            db['clientes'].append(cliente.to_dictionary())
            if(save_database()):
                print("\n** Cliente registrado **\n")
            else:
                print("\n** Error al registrar estudiante **\n")
        else:
            print("\n** Error al registrar estudiante **")
    except IndexError:
        print("\n** Error informacion insuficiente**")
    
def existe_cliente(codigo_cliente):
    clientes=db['clientes']
    index=-1
    for cliente in clientes:
        if (cliente['numero_socio']==codigo_cliente):
            index=clientes.index(cliente)
            break
    return index

def find_client(codigo_cliente):
    clientes=db['clientes']
    index=-1
    for cliente in clientes:
        if (cliente['numero_socio']==codigo_cliente):
            print("- Código: ",cliente['numero_socio'],
                  "\tNombre: ",cliente['nombre'],
                  "\tApellido: ",cliente['apellido'],
                  "\tDireccion: ",cliente['direccion'],
                  "\tCuentas Disponibles: ",cliente['cuentas_disponibles'])
            index=clientes.index(cliente)
            break
    if (index==-1):
        print("***No se encontró al estudiante en la base de datos***")
    return index
    
def delete_client():
    codigo_cliente=input("\nIngrese el código del cliente a eliminar:\n")
    index=existe_cliente(codigo_cliente)
    if(index!=-1):
        del db['clientes'][index]
        if(save_database()):
            print("\n** Cliente eliminado **\n")
        else:
            print("\n** Error de eliminacion**\n")
    else:
        print("** Error de eliminacion**")
    
    
def update_client():
    cliente_string=input("\nIngrese el código del cliente a modificar y sus nuevos datos:\n")
    cliente_array=cliente_string.split(',')
    try:
        cliente=Cliente(cliente_array[0],cliente_array[1],cliente_array[2],cliente_array[3],cliente_array[4])
        index=existe_cliente(cliente.numero_socio)
        if(index!=-1):
            db['clientes'][index]=cliente.to_dictionary()
            if(save_database()):
                print("\n** Cliente actualizado **\n")
            else:
                print("\n** No se actualizo cliente**\n")
        else:
            print("** Error actualizando cliente**")
    except IndexError:
        print("\n** Error actualizando cliente**")
```
